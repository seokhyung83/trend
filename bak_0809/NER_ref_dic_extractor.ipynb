```python
import pandas as pd
import openpyxl
from konlpy.tag import Mecab
mecab=Mecab()
```

```python
tmp_ref_dic = openpyxl.load_workbook('210803_라벨 데이터_v2.7.xlsx')
```

```python
sheet_name = ['part1', '스펙_키워드한정']
```

```python
tmp_dic = pd.DataFrame(tmp_ref_dic[sheet_name[0]].values).copy()
tmp_spec= pd.DataFrame(tmp_ref_dic[sheet_name[1]].values).copy()
```

```python
spec_tab = tmp_spec.iloc[2:,:]
spec_tab.columns = tmp_spec.iloc[1,:]
spec_tab = spec_tab.reset_index(drop=True)
spec_tab
```

```python
dic_tab = tmp_dic.iloc[1:, :15]
dic_tab.columns = tmp_dic.iloc[0,:15]
dic_tab = dic_tab.reset_index(drop=True)
dic_tab.head()
```

```python
add_df = pd.DataFrame()
temp = []
for i, row in  dic_tab.iterrows():
    for no in ['01','02','03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
        #s_temp = []
        if row['동의어/유사어_{}'.format(no)] is not None:
            temp.append([row['동의어/유사어_{}'.format(no)], row['대분류'], row['중분류']])
            #s_temp.append(row['동의어/유사어_{}'.format(no)])
            #s_temp.append(row['대분류'])
            #s_temp.append(row['중분류'])
```

```python
new_ref_dic = pd.DataFrame(temp, columns = ['명칭', '대분류', '중분류'])
```

```python
final_ref_dic = pd.concat([dic_tab[['명칭','대분류','중분류']], new_ref_dic])
final_ref_dic.to_csv('ref_dic.csv', encoding='utf-8-sig', index=False)
```
