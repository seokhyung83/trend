# 단어 추가

```python
load_dic = pd.DataFrame()
for i in range(20):
    try: 
        ref_dic = pd.read_excel('210803_라벨 데이터_v2.7.xlsx', i)
        load_dic=pd.concat([load_dic, ref_dic])
    except ValueError:
        pass
```

```python
load_dic
```

```python
ref_dic = load_dic.fillna('null')
```

```python
ref_dic.head(1)
```

```python
add_df = pd.DataFrame()
temp = []
for i,row in ref_dic.iterrows():
    
    for no in ['01','02','03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
        s_temp = []
        if row['동의어/유사어_{}'.format(no)] !='null':
            s_temp.append(row['동의어/유사어_{}'.format(no)])
            s_temp.append(row['대분류'])
            s_temp.append(row['중분류'])

        if len(s_temp) == 3:
            temp.append(s_temp)

#         temp.append(list(zip(row['동의어/유사어_01'], row['대분류'], row['중분류'])))
```

```python
temp
```

```python
new_ref_dic = pd.DataFrame(temp, columns = ['명칭', '대분류', '중분류'])
```

```python
new_ref_dic
```

```python
final_ref_dic = pd.concat([ref_dic[['명칭', '대분류', '중분류']], new_ref_dic])
```

```python
final_ref_dic.to_csv('ref_dic.csv', encoding='utf-8-sig', index=False)
```
